```python
import pandas as pd
import numpy as np
import re
from tqdm import tqdm
```

```python
dfa = pd.read_csv('지식인3_라벨링_코더A.csv')
dfb = pd.read_csv('지식인3_라벨링_코더B.csv')
dfc = pd.read_csv('지식인3_라벨링_코더C.csv')
```

```python
df = pd.concat([dfa, dfb, dfc], axis=1)
print(df.shape)
df.head()
```

```python
df.columns
```

```python
df = df.iloc[:,[0,1,7,13,14]]
print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset='review', inplace=True)
```

```python
df[df.코더A.isnull()]
```

```python
df['코더A'] = df.코더A.fillna(0)
```

```python
df.isnull().sum()
```

```python
# mode 함수를 사용하여 다수결로 target 컬럼 생성
df['target'] = df[['코더A', '코더B', '코더C']].mode(axis=1)[0].astype(int)

df.head()
```

```python
df
```

```python
df_doc = df.review
```

```python
def preprocess_text(text):
    if isinstance(text, str):
        text = re.sub(r'\n', ' ', text) # 줄넘김 없앰
        text = re.sub(r'\s+', ' ', text) # 공백 두 개 이상은 공백 하나로
        text = re.sub(r'[^\w\s]', ' ', text) # 특수문자 제거
        text = re.sub(r'http[s]?://\S+', ' ', text) #re.sub(r'https?://\S+', '', text) url 제거
        text = re.sub(r'\d', ' ', text) # 숫자 제거

    return text
```

```python
df_doc = df_doc.apply(preprocess_text)
df_doc
```

```python
df.review = df_doc
```

```python
df.to_csv('지식인3_코더통합.csv', encoding='utf-8-sig')
```
